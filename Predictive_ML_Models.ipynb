The latest advancements in training predictive machine learning models center on foundation forecasting models, edge machine learning, and hybrid AI. MLOps pipelines now natively integrate explainability dashboards (using frameworks like SHAP or LIME) so stakeholders can audit, trace, and understand exactly which features drive a model's prediction.

- **Automated Machine Learning (AutoML)**: Platforms are leveraging meta-learning and reinforcement learning to automate feature engineering, model selection, and hyperparameter tuning, allowing teams to scale their predictive projects much faster.

## Dataset with Mixed Data Type (Data Engineering)

To handle a dataset with mixed data types, follow a structured pipeline from data preparation to model selection.

### Overview

1. Data Type Categorization:
  
    First, group your features into three main buckets:
    - Numerical: Decimal values, float numbers, and integers (e.g., income, temperature).
    - Categorical: Text or ID labels with multiple distinct values (e.g., city, product type).
    - Binary: True/False, 1/0, or Yes/No flags (e.g., is_member, has_churned).
    - Long Descriptions: Generate embeddings

2. The Starting Point: Baseline Models

    Always start with a Tree-Based baseline before trying Neural Networks. They require minimal data transformation.
    - XGBoost, LightGBM, or CatBoost: These algorithms handle mixed data types natively.
    - Why start here? They provide a fast, high-performance benchmark and highlight feature importance.

3. Data Preprocessing Pipeline

    If you decide to move to Neural Networks, you must transform your data first:
    - For Numerical/Float Data: Apply Standardization (subtract mean, divide by standard deviation) or MinMax scaling. Neural networks fail if numerical features are on vastly different scales.
    - For Binary Data: Keep them as 0 and 1. No scaling is needed.
    - For Categorical Data:
      - Use One-Hot Encoding if there are only a few categories (low cardinality).
      - Use an Embedding Layer if there are many categories (high cardinality). This maps labels to dense vectors.
      
4. Where to Go: Choosing the NN Architecture
    If your baseline isn't enough, transition to one of these network styles:
    - Multi-Layer Perceptron (MLP): The simplest approach. Concatenate your scaled numbers, binary flags, and categorical embeddings into one flat vector and pass it through standard dense layers.
    - TabNet: A specialized neural network architecture that uses sequential attention to select features, mimicking the way decision trees work.
    - FT-Transformer: A transformer-based model designed specifically for tabular data that excels at learning interactions between mixed data types.

### Descriptive Textual Data

Descriptive textual data—like police reports, witness statements, and claims adjuster notes—must be processed before a machine learning model can use them. Computers cannot understand raw text directly.

You do not necessarily have to turn them into simple discrete categories. Instead, you can convert them into numerical formats that capture their meaning.


1. The Preprocessing Phase (Cleaning)
    
    Before converting text to numbers, you must clean it to remove "noise." This is typically done using Python (with libraries like nltk or spaacy) or cloud tools before the data feeds the model:
    - Lowercasing: Convert all text to lowercase so "Accident" and "accident" are treated the same.
    - Removing Stop Words: Delete common words that carry no predictive value (e.g., "the", "is", "at").
    - Punctuation & Special Characters: Strip out commas, hashes, and formatting glitches from document parsing.
    - Lemmatization: Reduce words to their base form (e.g., "crashing", "crashed", and "crashes" all become "crash").

2. Processing Strategy A: Transforming Text into Categories

    For certain specific text fields, turning them into discrete categories (called Categorical Variables) works best.
    - How it works: You use keyword matching or Regex to search the text for specific flags.
    - Example: If an adjuster's note contains the words "airbag deployed" or "totaled", you create a discrete binary category: `Airbag_Deployed = 1` or `Vehicle_Totaled = 1`.
    - Encoding: Once categorized, these are converted to numbers using One-Hot Encoding (turning categories into columns of 0s and 1s) inside BigQuery ML.

3. Processing Strategy B: Extracting Sentiment and Metrics

    Instead of categories, you can extract numerical scores from the text.
    - Sentiment Analysis: Convert a witness statement into a numerical "sentiment score" between -1 (highly negative/aggressive) and +1 (positive). In fraud modeling, highly erratic or overly defensive text sentiment can be a strong predictor.
    - Entity Counting: Count the number of specific high-risk entities mentioned (e.g., number of medical terms used).

4. Processing Strategy C: Text Embeddings (Advanced & Recommended)

    Modern machine learning rarely relies on just categories for text. Instead, we use Text Embeddings.
    - How it works: You pass the cleaned text through a Large Language Model (like Google's Vertex AI embedding models integrated with BigQuery).
    - The Output: The model transforms a long paragraph of text into a fixed list of numbers (a vector, such as 768 numbers).
    - Why it is powerful: These numbers capture the semantic meaning. A text saying "Vehicle smashed into a tree" and "Car collided with an oak" will generate very similar mathematical vectors, allowing your predictive model to understand they mean the same thing.

### How to do it in BigQuery

If you want to keep your pipeline entirely inside BigQuery, you can use BigQuery ML's native text functions or connection to Vertex AI:

- `ML.TRANSFORM`: Allows you to automatically tokenize text and perform standard text processing during model training.
- `ML.GENERATE_TEXT_EMBEDDING`: Allows you to call Google's Vertex AI models directly via SQL to turn your adjuster notes into vector numbers.

**NOTE!** The more the binary files/documents are structured and definite, the better for the data quality and therefore more deterministic results i.e., imagine if instead of descriptions there were certain boxes to check - of course designed/organized in a more readable manner for vision OCR or parser if any used.

## ML Pipeline inside/outside BigQuery
Keeping the pipeline entirely inside BigQuery means using BigQuery ML (BQML) [bqml-introduction]. This allows you to train and deploy machine learning models using standard SQL directly where your data lives, without writing Python code or moving datasets [bqml-introduction]. Here is an explanation of that approach, followed by the external alternatives.

### **Approach 1**: Keeping the Pipeline Entirely Inside BigQuery (BigQuery ML)

When you keep everything inside BigQuery, the database itself acts as the training engine [bqml-introduction].
  
  How it Works:
  - Feature Engineering: You use standard SQL commands (`SELECT`, `JOIN`, `CASE WHEN`) to clean and prepare data.
  - Model Training: You use a SQL statement like `CREATE` OR `REPLACE MODEL` specifying your target variable and model type (e.g., Logistic Regression, XGBoost) [bqml-introduction].
  - Prediction: You run a SQL query using `ML.PREDICT` to get instant scores on new data [bqml-introduction].
  
  Advantages:
  - No Data Movement: You do not export massive insurance datasets, saving time and transfer costs.
  - Simplicity: Data analysts who only know SQL can build complex machine learning models [bqml-introduction].
  - Speed: It leverages Google Cloud's massive, scalable computing power automatically.

### **Approach 2**: Alternatives Outside BigQuery (The "Export" Method)

If BQML is too restrictive, the standard alternative is to pull the data out of BigQuery and pipe it into a dedicated machine learning environment.

1. Vertex AI Custom Training (Python/Notebook Environment)
  - What it is: You export BigQuery data into Python notebooks (Jupyter/Vertex AI Workbench) or custom Python scripts.
  - Why use it: You need complete freedom over your code. You can use advanced libraries like PyTorch, scikit-learn, or TensorFlow to build highly customized deep learning models.

2. Vertex AI AutoML
  - What it is: Google Cloud’s point-and-click automated machine learning service.
  - Why use it: You point AutoML directly at your BigQuery table. It automatically tests dozens of different model types, handles feature engineering, and picks the best model without you writing any ML code.
  
3. Third-Party Platforms (Databricks, Snowflake, AWS)
  - What it is: Moving the data completely out of Google Cloud into external ecosystems.
  - Why use it: Your company already uses Databricks (Spark) for heavy machine learning or uses AWS (SageMaker) for its infrastructure.

### Standard BigQuery ML (BQML) vs. BigQuery AutoML

**Standard BigQuery ML (BQML)**
- What it is: You manually select the model type (e.g., linear regression, boosted trees, random forest) and configure parameters yourself.
- Control: You have high control over feature engineering, model parameters, and hyperparameter tuning.
- Usage: Best for quick iterations, baselines, and when you require transparency into how the model weights work.

**BigQuery AutoML**
- What it is: You let Google's automated machine learning do the heavy lifting. It automatically tests multiple algorithms, tunes hyperparameters, and performs feature engineering.
- Control: Low control. You define the training budget (how many hours it runs), and the system finds the best model.
- Usage: Best when you want to achieve the highest possible predictive accuracy on tabular data without manual trial-and-error.

**How to Use AutoML in BigQuery**

You can train an AutoML model directly from the BigQuery console using a simple SQL statement:

```
CREATE MODEL my_dataset.my_automl_model
OPTIONS(model_type='AUTOML_CLASSIFIER', -- or AUTOML_REGRESSOR
        input_label_cols=['target_column'],
        budget_hours=1.0)
AS
SELECT * FROM my_dataset.training_data;

```

**Key Trade-offs to Consider**

- Time & Cost: AutoML training takes much longer (minimum 1 hour) and costs more in compute than a standard BQML boosted tree or linear regression.
- Explainability: While AutoML gives great metrics and high-level feature importance, you cannot inspect the exact model weights or underlying tree structures as you can with custom BQML models.

### Data Tranformation

You should never use the variables exactly as they are extracted from your documents. Raw extracted text (like a police report narrative) or unformatted fields (like localized dates) cannot be read by an ML algorithm. They must be transformed into numbers, matrices, or clean vectors before being injected into the pipeline. Here is exactly how that process works and how you can do it.

#### Part 1: Why Raw Extracted Data Fails

If you feed raw document data straight to an ML model, the training will fail or give inaccurate results because:

- **The "Text" Problem**: An ML model cannot read a string like "Vehicle sustained heavy front-end damage." It needs that converted into a numerical severity score or a category flag (1 or 0).
- **The "Scale" Problem**: If your `Claim_Amount` is $50,000 and the `Driver_Age` is 25, the model might mistakenly think the claim amount is 2,000 times more important simply because the number is larger. They must be scaled.
- **The "Missing" Problem**: OCR/Document AI tools frequently miss fields. An ML model cannot handle a blank space (NULL); it will crash unless you tell it how to fill that gap.

#### Part 2: How to Transform and Inject (The 2 Methods)

You have two choices for where and how this transformation happens: In-Database (SQL) or Pre-Pipeline (Python).

**Method A: The Modern Way (In-Database with BigQuery ML)**

Instead of creating a secondary, messy "transformed table," you write your cleaning rules directly inside the model definition using BigQuery’s `TRANSFORM` clause [bqml-introduction, 1.1.2]. BigQuery remembers these rules and applies them automatically every time new data passes through.

Here is how you write that SQL injection:

```
CREATE OR REPLACE MODEL `your_project.insurance_ml.predict_fraud`
TRANSFORM(
  -- 1. Scale numbers so large amounts don't break the math
  ML.STANDARD_SCALER(claim_amount) OVER() AS scaled_amount,

  -- 2. Fill missing age values automatically with the average age
  ML.IMPUTER(driver_age, 'mean') OVER() AS imputed_age,

  -- 3. Extract time context out of a raw timestamp text
  EXTRACT(HOUR FROM accident_timestamp) AS accident_hour,

  -- 4. Pass clean variables straight through (BQML automatically encodes strings)
  vehicle_make,
  
  -- This is the target variable we want to predict
  is_fraudulent
)
OPTIONS(
  model_type='LOGISTIC_REG_COEF', -- The ML algorithm
  input_label_cols=['is_fraudulent'] -- Telling the pipeline what to predict
) AS
SELECT * FROM `your_project.insurance_database.raw_extracted_claims`;
```

**IMPORTANT NOTE!** BigQuery ML can handle part of the preprocessing and feature engineering, but not all of what a dedicated Python/Vertex AI pipeline can do.

What BigQuery ML can do: Basic preprocessing
- Missing value handling (some models)
- Feature scaling/normalization (for certain algorithms)
- Train/test splitting
- Automatic encoding of categorical variables
- Text processing for some text models

Examples:

```
CREATE MODEL ...
OPTIONS(model_type='logistic_reg')
AS
SELECT
  age,
  vehicle_type,
  claim_amount,
  target
FROM claims;
```
BigQuery ML can automatically encode `vehicle_type`.

Example: SQL-based feature engineering

You can create features directly in SQL.

```
SELECT
  claim_amount,
  DATE_DIFF(claim_date, accident_date, DAY) AS days_to_report,
  LOG(claim_amount) AS log_claim_amount
FROM claims;
```

Having said all the above, BigQuery ML is NOT ideal for advanced feature engineering.

**Method B: The Traditional Way (Python & Dataframe Ingestion)**

If you prefer building models in Python (Vertex AI, PyTorch, or Scikit-Learn), you leave your BigQuery data raw, pull it into a notebook, and transform it there before passing it to your model.

1. Pull Raw Data: Use the `bigframes` library to stream your BigQuery table directly into a Python notebook environment without downloading heavy CSVs.
2. Transform with Code: Use Python libraries to map out your rules:
    - Use `scikit-learn`'s `OneHotEncoder` to convert categorical text fields into binary columns.
    - Use `StandardScaler` to clean up numerical inputs.
3. Inject: Pass the resulting NumPy array or Pandas Dataframe directly into your model's `.fit()` training command.

**QUESTION**: Which approach should you use?

Go with Method A (BigQuery ML) if you want the easiest path. By using the TRANSFORM clause inside BigQuery, you don't have to manage two separate data pipelines (one for training and one for incoming production claims). The database handles the transition seamlessly.

### Multi-layer Architecture

In modern cloud data architectures, you never destroy your raw data. Instead, you use a multi-layered approach within BigQuery to keep both safe and separated. Here is exactly how the data flows and where it is stored:

**The Architecture: Multi-Layer Storage**

Instead of a single database that replaces raw data with transformed data, your BigQuery environment will typically look like this:

Layer 1: The Raw Landing Zone (Bronze Layer)
- What it is: A secure dataset containing the exact, unedited variables extracted from your documents (OCR outputs, raw timestamps, unstructured notes).
- Access: Restricted to automated ingestion pipelines and data engineers.
- Rule: This layer is read-only and acts as your historical "Source of Truth." You never delete it. If your ML model fails or you want to build a completely new model next year, you will need to re-read this raw data.

Layer 2: The Transformed View/Table (Silver Layer)
- What it is: This is where your cleaned data lives. You do not manually overwrite tables. Instead, you create a SQL View or a Scheduled Query that reads the raw data, applies the transformations, and outputs the result.
- How it works: It acts as a wrapper. When the raw data updates, the transformed layer updates automatically.

Layer 3: The Model Artifact (Gold Layer)
- What it is: If you use BigQuery ML (BQML), the transformation logic is saved inside the Model Object itself [bqml-introduction, 1.1.2].
- How it works: BigQuery reads from Layer 1, processes the math on the fly, trains the model, and stores the rules inside the model [bqml-introduction]. The transformed data is processed in-memory and does not need to be saved as a permanent, massive physical table unless you explicitly want to.

**Visualizing the Pipeline Flow**
```
[Raw Documents]
       │ (Document AI / Extraction)
       ▼
┌──────────────────────────────────────┐
│  Layer 1: Raw BigQuery Table         │ <── NEVER DELETED (Your permanent record)
└──────────────────────────────────────┘
       │
       │ (SQL Transformation / BQML)
       ▼
┌──────────────────────────────────────┐
│  Layer 2: Transformed View           │ <── Virtual layer for the ML pipeline
└──────────────────────────────────────┘
       │
       ▼
┌──────────────────────────────────────┐
│  Layer 3: Trained ML Model           │ <── Stores both the math and the rules
└──────────────────────────────────────┘
```

**QUESTION**: Why is keeping the raw data critical for Insurance ML?

1. Compliance and Auditing: If a claimant disputes a rejected claim, fraud investigators and legal teams must be able to see the exact text extracted from the original police report, not just a scaled numerical value (like 0.42) used by the machine learning model.
2. Model Upgrades: If you decide to switch from a simple model to an advanced Deep Learning model in six months, that new model will require raw, unmanipulated data to train effectively.

Because training an ML model requires scanning millions of rows over and over again, you must balance SQL complexity, compute costs, and data storage. Therefore, the choice between a Standard View, a Materialized View, or a Scheduled Query matters immensely. For a machine learning pipeline, picking the wrong one can result in syntax crashes, sky-high cloud bills, or sluggish model training.

#### Standard (Logical) Views

A standard view is purely a saved virtual query. It stores zero physical data on disk.

- How it behaves with ML: Every single time your ML pipeline or model runs an epoch, hyperparameter tuning job, or training query, BigQuery executes the entire underlying transformation SQL from scratch.
- The Problem: If your transformation contains heavy text parsing or complex regex to clean insurance document text, you will pay full compute costs every single time the model references the view.
- Best Use Case: Excellent for prototyping or light feature adjustments. Avoid it for final, massive ML training runs.

#### Materialized Views

Materialized views are precomputed caches that physically save data to disk and automatically auto-update when the raw data changes.

- The Problem (CRITICAL FOR ML): BigQuery places strict SQL syntax limitations on materialized views. They do not support `UNNEST` functions (needed for parsing document arrays), analytic/window functions, `COUNT DISTINCT`, or complex multi-table `JOIN` statements.
- Best Use Case: Do not use them for your primary feature engineering layer. They are too restrictive for advanced ML data preparation.

#### Scheduled Queries (The Recommended Approach)

A scheduled query runs a custom SQL script on a timer (e.g., every night at 1:00 AM) and writes the results into a permanent physical BigQuery table.

- How it behaves with ML: Your ML pipeline points directly to a flat, physically optimized table. When the model trains, it reads pre-calculated data instantly, bypassing all execution overhead.
- Why it wins: It supports 100% of all SQL syntax (complex regex, array flattening, window functions, and external AI calls). You pay the heavy compute cost only once when the scheduled query runs, rather than paying it repeatedly during model training iterations

To learn more about BigQuery Views you can also visit [Google Cloud's official documentation](https://docs.cloud.google.com/bigquery/docs/views-intro).

### How Views Are Used: Big Query ML vs. External ML Tools

Views are heavily utilized as virtual, abstract layers to implement machine learning pipelines both natively inside BigQuery ML (BQML) and when exporting data to non-BigQuery external tools. Because logical views in BigQuery are essentially stored SQL queries that execute dynamically when called, they serve as a flexible and reliable mechanism for data preparation and consumption across various ML architectures.

1. How Views Are Used in BigQuery ML (BQML) \
When you build an ML pipeline entirely inside BigQuery using SQL, views are vital for structuring the pipeline stages without creating duplicate physical tables:
    - Feature Engineering: You can encapsulate complex data transformations—such as joins, aggregations, and mathematical treatments—into a view. This view serves as the clean feature matrix.
    - Model Training Ingestion: Instead of pointing your training query to raw tables, you pass the view directly into the CREATE MODEL statement (e.g., CREATE MODEL my_dataset.my_model AS SELECT * FROM my_dataset.my_clean_features_view).
    - Decoupled Data Splits: You can use views to easily partition your datasets into dynamic train, validation, and test subsets without modifying underlying records.
    - Inference Input: When generating batch predictions via ML.PREDICT, you can use a view to feed newly arrived, un-scoured production data directly into the model.

2. How Views Are Used with External ML Tools \
When training models externally—such as in Vertex AI, Databricks, Snowflake, or local Python environments (Jupyter, PyTorch, Scikit-learn)—views act as the standard interface for data extraction:
    - Feature Store Integration: External platforms like Vertex AI Feature Store point directly to BigQuery views (or "Feature Views") to manage offline data extraction for training or low-latency online serving.
    - Data Abstraction & Security: External tools do not need permission to view raw, sensitive, or messy tables. Data engineers create an Authorized View that restricts columns and exposes only the precise, anonymized rows needed by data science teams.
    - Consistent Feature Definition: Pointing external pipelines to a view ensures that both your offline training code and your online production application query the exact same SQL logic, preventing training-serving skew.

**Summary: Logical vs. Materialized Views in Pipelines** \
Depending on your pipeline requirements, you can choose between two main view types:
- Logical Views: Great for data that changes constantly or when storage savings are prioritized, though they compute on-the-fly and incur processing costs each run.
- Materialized Views: BigQuery caches the query results of Materialized Views for higher performance and lower costs. This is highly advantageous for massive datasets that need to be repeatedly evaluated by automated ML retraining loops.

While views are highly recommended for simplicity, more sophisticated and efficient alternatives exist for complex, production-grade ML pipelines.

Views have a major performance drawback: standard logical views re-run their underlying SQL query from scratch every time they are called [VIEWS-INTRO]. For massive data or frequent machine learning (ML) training loops, this introduces heavy compute latency and high query costs.

Modern data architecture offers several advanced alternatives that balance efficiency, automation, and speed.

### Orchestration Tables via dbt or Apache Airflow (The Data Engineering Standard)

In mature data platforms, engineers use data transformation tools like dbt [Data Build Tool](https://www.getdbt.com/) to build "incremental tables" rather than views.

- Why it is better: dbt transforms raw data and materializes it into a physical, static table inside BigQuery on a scheduled cadence (e.g., every night or every hour).
- The ML Benefit: When an ML tool or BigQuery ML goes to train a model, it queries a static, fully indexed physical table [BQML-INTRODUCTION]. This results in blazing-fast data ingestion speeds and minimal BigQuery processing costs during model training.

**Summary Recommendation**

- Use Logical Views if you are just starting out, testing features, or working with datasets under a few gigabytes [VIEWS-INTRO].
- Upgrade to Materialized Views if your ML model relies heavily on aggregations (like averages or sums over time) and query costs are rising [MATERIALIZED-VIEWS-USE].
- Transition to dbt-managed tables or a Feature Store if you are building an enterprise-grade, automated production pipeline where pipeline reliability and data lineage are strictly required.
